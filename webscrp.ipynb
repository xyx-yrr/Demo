```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://webscraper.io/test-sites/e-commerce/static"
```

```python
response = requests.get(url)
print ( "status code:",response.status_code)
```

```python
soup = BeautifulSoup(response.text,("html.parser"))
print ("WEBPAGE PARSED SUCCESSFULLY")
```

```python
products = soup.find_all('div',class_='thumbnail')
print ("Total Products:",len(products))
```

```python
# Creating empty lists

product_name_list = []
price_list = []
review_list = []
description_list = []
rating_list = []
```

```python
for product in products:
    product_name = product.find('a',class_='title').text.strip()
    price = product.find('h4',class_='price').text.strip()
    description = product.find('p',class_='description').text.strip()
    reviews = product.find('p',class_='pull-right')
    if reviews:
        review_tag=reviews.text.strip()
    else:
        review_tag='NO REVIEWS'
    rating_tag = product.find('div',class_='ratings')
    if rating_tag:
        ratings = len(rating_tag.find_all('span',class_='glyphicon glyphicon-star'))
    else:
        ratings=0
```

```python
    product_name_list.append(product_name)

    price_list.append(price)

    description_list.append(description)

    review_list.append(reviews)

    rating_list.append(ratings)
```
